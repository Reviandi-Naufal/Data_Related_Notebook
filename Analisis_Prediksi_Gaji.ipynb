```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()

train = pd.read_csv('Pendapatan.csv')
test = pd.read_csv('Pendapatan_test.csv')
```

```python
train.head(50)
```

```python
test.head(50)
```

```python
train_test_data = [train, test]

for dataset in train_test_data:
    dataset.replace('?', np.nan, inplace=True)
```

```python
train.isna().sum()
```

```python
test.isna().sum()
```

```python
def bar_chart(feature):
    lebih_5jt = train[train['Gaji']==1][feature].value_counts()
    kurang_5jt = train[train['Gaji']==0][feature].value_counts()
    df = pd.DataFrame([lebih_5jt,kurang_5jt])
    df.index = ['lebih_5jt','kurang_5jt']
    df.plot(kind='bar',stacked=True, figsize=(10,5))
```

```python
bar_chart('Status Perkawinan')
```

Grafik menunjukkan orang yang sudah menikah lebih mungkin memiliki pendapatan lebih dari 5jt dan yang belum pernah menikah lebih mungkin memiliki pendapatan kurang dari 5jt.

```python
bar_chart('Jenis Kelamin')
```

Grafik menunjukkan laki2 lebih mungkin mendapatkan gaji lebih dari 5 juta namun lebih mungkin juga mendapatkan gaji karng dari 5 jt. Sementara wanita lebih mungkin mendapatkan gaji dibawah 5jt.

## 1. Feature Engineering dan Data Cleaning

### 1.1 Jenis Kelamin

```python
sex_mapping = {"Laki2": 0, "Perempuan": 1}
for dataset in train_test_data:
    dataset['Jenis Kelamin'] = dataset['Jenis Kelamin'].map(sex_mapping)
```
